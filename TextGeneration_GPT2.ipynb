#Task-01: Text Generation using GPT-2

This project demonstrates how to generate text using a pretrained GPT-2 model from Hugging Face Transformers.

We explore three generation techniques:
1. Greedy Search
2. Beam Search
3. Sampling (Top-k and Top-p)

No dataset is required since we use a pretrained language model.


```python
!pip install transformers torch
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
prompt = "Artificial Intelligence"
inputs = tokenizer(prompt, return_tensors="pt")

#Generate outputs first
output_greedy = model.generate(**inputs, max_new_tokens=150)

output_beam = model.generate(**inputs, num_beams=5,
    early_stopping=True,
    max_new_tokens=150)

output_sample = model.generate(**inputs,do_sample=True,
    top_k=50,
    top_p=0.95,
    max_new_tokens=150)

# Display nicely
from IPython.display import display, Markdown

display(Markdown("## GREEDY SEARCH OUTPUT"))
display(Markdown(tokenizer.decode(output_greedy[0], skip_special_tokens=True)))

display(Markdown("---"))

display(Markdown("## BEAM SEARCH OUTPUT"))
display(Markdown(tokenizer.decode(output_beam[0], skip_special_tokens=True)))

display(Markdown("---"))

display(Markdown("## SAMPLING OUTPUT"))
display(Markdown(tokenizer.decode(output_sample[0], skip_special_tokens=True)))
```

## Analysis of Results

- Greedy Search produces repetitive text because it always selects the highest probability token.
- Beam Search improves structure but may still generate repeated phrases.
- Sampling (Top-k, Top-p) produces more creative and diverse outputs.

Sampling is generally preferred for creative text generation tasks.
